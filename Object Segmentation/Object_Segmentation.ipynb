```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the input image
input_image = cv2.imread("image2.jpg")
```

```python
#Convert the image to grayscale
gray_img = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
```

### Original image

```python
# Display the original image and the extracted objects
plt.figure(figsize=(15, 10))
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis('off')
```

### Simple thresholding

```python
# Simple thresholding
_, simple_thresholded_image = cv2.threshold(gray_img, 127, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 2)
plt.imshow(simple_thresholded_image, cmap='gray')
plt.title("Technique1: Simple Thresholding")
plt.axis('off')
```

### Adaptive thresholding

```python
# Adaptive thresholding
adaptive_thresholded_image = cv2.adaptiveThreshold(gray_img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 2)
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 3)
plt.imshow(adaptive_thresholded_image, cmap='gray')
plt.title("Technique2: Adaptive Thresholding")
plt.axis('off')
```

### Otsu's thresholding

```python
# Otsu's thresholding
_, otsu_thresholded_image = cv2.threshold(gray_image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 4)
plt.imshow(otsu_thresholded_image, cmap='gray')
plt.title("Technique3: Otsu's Thresholding")
plt.axis('off')
```

### Canny edge detection

```python
# Canny edge detection
canny_edges = cv2.Canny(gray_image, 100, 200)
```

```python
plt.figure(figsize=(18, 12))
plt.subplot(2, 3, 5)
plt.imshow(canny_edges, cmap='gray')
plt.title("Technique4: Canny Edge Detection")
plt.axis('off')

plt.show()
```
