# Create an Audiobook from a PDF
## This task tests your ability to apply Text to Speech conversion and Extraction of Text from PDF files in the creation of an audiobook from a PDF file

### Steps

- Clean the text

- Save the speech
- Play the speech

## 1. Extract text from PDF
- Use PyPDF2

### Install the library

```python
!pip install PyPDF2
```

### Import the library

```python
import PyPDF2
```

### Extract the text

```python
# Reading the PDF File
pdf_file_path = 'Neil_armstrong_wikipedia.pdf'
from PyPDF2 import PdfReader
# Open the PDF file

with open(pdf_file_path, 'rb') as pdf_file:
    # Create a PDF file reader object
    pdf_reader = PdfReader(pdf_file)

    # Extract text from each page of the PDF file
    text = ''
    for page_num in range(len(pdf_reader.pages)):
        page = pdf_reader.pages[page_num]
        text += page.extract_text()   
```

### Print the extracted text

```python
print(text)
```

## 2. Convert the Text into Speech
- Use **pyttsx3** OR **gTTS**

### Install the library

### Import the library

```python
from gtts import gTTS
import os
import wave
```

### Initialize a Speaker object

```python
# Function to convert text to speech
def text_to_speech(text, filename):
    tts = gTTS(text=text, lang='en')
    tts.save(filename)
```

### Convert the text

```python
# Convert the extracted text to speech
output_file = 'output.wav'
text_to_speech(text, output_file)
```

### Save the audio

```python
print(f'Text converted to speech and saved as "{output_file}"')
```
